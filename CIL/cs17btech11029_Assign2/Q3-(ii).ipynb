```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
# (a) P(X,Y) = P(Y|X)P(X)

x = np.random.randn(100)
y = 4*x + np.random.randn(100)
d = {'x': x, 'y': y}

df = pd.DataFrame(data=d)

plt.figure()
sns.displot(df, x='x', y='y', kind="kde")
plt.show()
```

```python
# (b) P(Y|X=2)

y = 4*2 + np.random.randn(100)
plt.figure()
sns.distplot(y)
plt.show()
```

```python
# (c) P(Y|do(X=2))

y = 4*2 + np.random.randn(100)
plt.figure()
sns.distplot(y)
plt.show()
```
